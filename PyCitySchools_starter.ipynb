```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of schools
Total_schools = school_data_complete["school_name"].count()

#Calculate the total number of students
Total_students = school_data_complete["Student ID"].count()

#Calculate the total budget
Total_budget = school_data_complete["budget"].sum()

#Calculate the average math score
Math_score = school_data_complete["math_score"].mean()

#Calculate the average reading score
Reading_score = school_data_complete["reading_score"].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
Passing_math_score = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]

#Calculate the percentage of students with a passing reading score (70 or greater)
Passing_reading_score = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]

#Calculate the percentage of students who passed math and reading (% Overall Passing)
Passed_math_percentage = Passing_math_score / float(Total_students) * 100
Passed_reading_percentage = Passing_reading_score / float(Total_students) * 100

#Create a dataframe to hold the above results
District_summary = pd.DataFrame({"Total_schools": [Total_schools], "Total_students" : [Total_students], "Total_budget" : [Total_budget],
                                "Math_score" : [Math_score], "Reading_score" : [Reading_score],
                                "Average math score": [Passing_math_score],
                                "Average reading score": [Passing_reading_score],
                                 "Passed_math_percentage": [Passed_math_percentage],
                                "Passed_reading_percentage":[Passed_reading_percentage]})
District_summary
```

## School Summary

```python
school_data_complete
```

```python
#School Name
school_name = school_data_complete.get

#School type
school_type = school_data_complete["type"]

#Total students
total_student = school_data_complete["Student ID"].count()

#Total School Budget
Total_budget = school_data_complete["budget"].sum()

#Per Student Budget
Student_budget = Total_budget / total_student

#Average Math Score
Average_math_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]

#Average Reading Score
Average_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

#% Passing Math
School_passing_math = school_data_complete[(school_data_complete["math_score"] >=70)]

#% Passing Reading
School_passing_reading = school_data_complete[(school_data_complete["reading_score"] >=70)]

# % Passing (The percentage of students that passed math and reading.)
passing_rate_math = School_passing_math.groupby(["school_name"]).count()["student_name"] / total_student * 100
passing_rate_reading = School_passing_reading.groupby(["school_name"]).count()["student_name"] / total_student * 100
overall_passing_rate = (passing_rate_math + passing_rate_reading) / 2

#Create a dataframe to hold the above results
School_summary = pd.DataFrame({"School Type": [school_type], 
                               "Total Students" : [total_student], 
                               "Total School Budget" : [Total_budget],
                                "Per Student Budget" : [Student_budget], 
                               "Average Math Score" : [Average_math_score],
                                "Average Reading Score": [Average_reading_score],
                                "% Passing Math": [School_passing_math],
                                 "% Passing Reading": [School_passing_reading],
                                "% Overall Passing Rate":[overall_passing_rate]})
School_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Establish the bins 
bins = [0, 100, 200, 300, 400]
group_names = ["<$100", "$100-200", "$200-300", "$300-400"]

# Categorize the spending based on the bins
School_summary["spend Ranges (Per Student)"] = pd.cut(Student_budget, bins, labels=group_names)

spend_math_scores = School_summary.groupby(["spend Ranges (Per Student)"]).mean()["Average math score"]
spend_reading_scores = School_summary.groupby(["spend Ranges (Per Student)"]).mean()["Average reading score"]
spend_passing_math = School_summary.groupby(["spend Ranges (Per Student)"]).mean()["% Passing Math"]
spend_passing_reading = School_summary.groupby(["spend Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_rate = (spend_passing_math + spend_passing_reading) / 2

# Assemble into data frame
spend_summary = pd.DataFrame({"Average Math Score" : spend_math_scores,
                                 "Average Reading Score": spend_reading_scores,
                                 "% Passing Math": spend_passing_math,
                                 "% Passing Reading": spend_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})


# Display results
spend_summary
```
